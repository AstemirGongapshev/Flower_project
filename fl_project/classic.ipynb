```python
import pandas as pd 
import json
import matplotlib.pyplot as plt 
import os 
```

```python
res_path = "prox_metrics.json"
res_path_fed_avg = "fed_avg_metrics.json"
```

```python
with open(res_path, 'r') as f:
    results = json.load(f)

with open(res_path_fed_avg, 'r') as f:
    results_fed_avg = json.load(f)
```

```python
plt.title("Accuracy")
pd.Series([results["Accuracy"][i][1] for i in range(len(results["Accuracy"]))]).plot()
plt.grid()
plt.show()
```

```python
plt.title("Loss")
pd.Series([results["Loss"][i][1] for i in range(len(results["Loss"]))]).plot()
plt.grid()
plt.show()
```

```python
plt.title("FED_PROX_ROC_AUC")
pd.Series([results["ROC_AUC"][i][1] for i in range(len(results["ROC_AUC"]))]).plot()
plt.grid()
plt.ylim(0.65, 0.8)
plt.show()
```

```python
plt.title("F1_Score")
pd.Series([results["F1_Score"][i][1] for i in range(len(results["F1_Score"]))]).plot()
plt.grid()
plt.show()
```
